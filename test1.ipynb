```python
# 列印專案結構
import os

def print_tree(start_path, max_depth=1, prefix="", ignore_list=None):
    if ignore_list is None:
        ignore_list = []

    start_path = os.path.abspath(start_path)
    base_name = os.path.basename(start_path)
    print(base_name + "/")
    
    def _walk(path, depth, prefix):
        if depth > max_depth:
            return
        items = sorted(os.listdir(path))
        # 過濾要忽略的項目
        items = [item for item in items if item not in ignore_list]

        for i, item in enumerate(items):
            full_path = os.path.join(path, item)
            connector = "└── " if i == len(items)-1 else "├── "
            print(prefix + connector + item)
            if os.path.isdir(full_path) and depth < max_depth:
                new_prefix = prefix + ("    " if i == len(items)-1 else "│   ")
                _walk(full_path, depth+1, new_prefix)

    _walk(start_path, 1, "")

# 使用範例：忽略 .venv 與 tests 資料夾
path = r"C:\Users\PARALELL\Desktop\code\Newfolder\11402\DocuMind_AI_lite_V2"
print_tree(path, max_depth=10,
           ignore_list=[".venv", "tests","__pycache__","cache","node_modules",".git","staging","tmp_yaml", "output"]) # 6 # "input", "output"
```
